```python
def age_group(age):
    """Create an age group for each patient using NACCDAGE"""
    if age in range (31, 36): #range excludes upper bound
        group = '31-35'
    if age in range (36,41):
        group = '36-40'
    if age in range (41,46):
        group = '41-45'
    if age in range (46,51):
        group = '46-50'
    if age in range (51,56):
        group = '51-55'
    if age in range (56,61):
        group = '56-60'
    if age in range (61,66):
        group = '61-65'
    if age in range (66,71):
        group = '66-70'
    if age in range (71,76):
        group = '71-75'
    if age in range (76,81):
        group = '76-80'
    if age in range (81,86):
        group = '81-85'
    if age in range (86,91):
        group = '86-90'
    if age in range (91,96):
        group = '91-95'
    if age in range (96,101):
        group = '96-100'
    if age in range (101,106):
        group = '101-105'
    if age in range (106,111):
        group = '106-110'
    
    return group
    
```

```python
def matplot_hist(data, histtype,label):
    '''--- plot the histogram of a given data----
    Function needs three arguments
    data, histtype = 'bar', 'barstacked', 'step', 'stepfilled'
    label = column_name'''
    import matplotlib.pyplot as plt
   
    plt.hist(data,histtype = histtype)
    plt.xlabel('bins')
    plt.ylabel('Frequency')
    plt.title(label)
    plt.figure(figsize=(10,10))
    plt.show()

def QQplot(data, line, label):
    ''' --- plot the Quantile-Quantile Plot of a given data
    str {45, s, r, q} or None
    Options for the reference line to which the data is compared:45
    - 45-degree line s - standardized line, 
    the expected order statistics are scaled by the standard deviation 
    of the given sample and have the mean added to them.
    r - A regression line is fit q - A line is fit through the quartiles.'''
    
    from statsmodels.graphics.gofplots import qqplot
    from matplotlib import pyplot
    qqplot(data = data, line = line)
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.title(label)
    plt.figure(figsize=(10,10))
    plt.show()
```

Building a linear regresion model (ordinary least square) for WHole_brain_weight against categorical value Age and Sex


```python
# fit a  linear model, C for categorical, ols = ordinary least square
formula = 'Whole_brain_weight ~ C(Age) + C(Sex)+ C(Age)*C(Sex)'
wholebrainwt_lm = ols(formula, data1).fit()
print(wholebrainwt_lm.summary())
```

```python
#ANOVA
wholebrainwt_table = sm.stats.anova_lm(wholebrainwt_lm, type = 1)
print(wholebrainwt_table)
```

```python
wholebrainwt_table = sm.stats.anova_lm(wholebrainwt_lm, type = 2)
print(wholebrainwt_table)
```
